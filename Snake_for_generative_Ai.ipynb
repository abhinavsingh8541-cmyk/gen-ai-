```python
def sum_number(*args):
  sum = 0
  for num in args:
    sum+=num
  return sum
```

```python
sum_number(3, 8, 9, 7, 6)
```

```python
def multiplication_number(*args):
  multiplication = 1
  for num in args:
    multiplication*=num
  return multiplication
```

```python
multiplication_number(3,4,5,6,8)
```

```python
#keyyword variable length argument
def test(**kwargs):
  print(kwargs)
  print(type(kwargs))
  print(len(kwargs))
```

```python
test(a = 10, b = 20, c= 30)
```

```python
def pass_student(**kwargs):
  pass_student_list =[]
  for name, marks, in kwargs.items():
    if marks >= 35:
      pass_student_list.append(name)
  return pass_student_list
```

```python
pass_student(A = 30, B = 40, C= 50, D = 80, E = 20)
```
